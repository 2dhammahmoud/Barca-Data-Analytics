```python
import json

# قراءة الملف الأصلي
with open('season_players_map.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# قاموس الأسعار بتاعك
TRANSFER_FEES = {
    # --- 1979-80 ---
    ('Simonsen', '1979-80'): '660k', ('Roberto Dinamite', '1979-80'): '300k', 
    ('Canito', '1979-80'): '270k', ('Landaburu', '1979-80'): '180k', ('Amigó', '1979-80'): '72k',
    
    # --- 1980-81 ---
    ('Bernhard Schuster', '1980-81'): '1.14M', ('Alexanko', '1980-81'): '720k', 
    ('Quini', '1980-81'): '480k', ('Amador', '1980-81'): '150k',
    
    # --- 1981-82 ---
    ('Víctor Muñoz', '1981-82'): '510k', ('Urruti', '1981-82'): '390k', 
    ('Morán', '1981-82'): '390k', ('Gerardo', '1981-82'): '270k', ('Cleo', '1981-82'): '90k',
    
    # --- 1982-83 ---
    ('Maradona', '1982-83'): '7.30M', ('Marcos Alonso', '1982-83'): '720k', 
    ('Julio Alberto', '1982-83'): '480k', ('Periko Alonso', '1982-83'): '420k', 
    ('Urbano', '1982-83'): '420k', ('Pichi Alonso', '1982-83'): '420k',
    
    # --- 1983-84 ---
    ('Gabrich', '1983-84'): '120k',
    
    # --- 1984-85 ---
    ('Archibald', '1984-85'): '2.00M',
    
    # --- 1985-86 ---
    ('Amarilla', '1985-86'): '300k',
    
    # --- 1986-87 ---
    ('Lineker', '1986-87'): '3.20M', ('Hughes', '1986-87'): '3.00M', 
    ('Zubizarreta', '1986-87'): '1.90M', ('Robert', '1986-87'): '750k',
    
    # --- 1988-89 ---
    ('López Rekarte', '1988-89'): '960k', ('Serna', '1988-89'): '1.25M', 
    ('Aloísio', '1988-89'): '1.20M', ('Eusebio', '1988-89'): '1.10M', 
    ('Bakero', '1988-89'): '1.38M', ('Begiristain', '1988-89'): '1.38M', 
    ('Julio Salinas', '1988-89'): '1.10M', ('Unzué', '1988-89'): '1.10M', 
    ('Soler', '1988-89'): '1.50M', ('Valverde', '1988-89'): '1.50M', 
    ('Manolo Hierro', '1988-89'): '900k', ('Romerito', '1988-89'): '240k'
}


# إضافة الأسعار
for season, players in data.items():
    for player_dict in players:
        player_name = player_dict['player']
        fee = TRANSFER_FEES.get((player_name, season))
        
        # معالجة الأسماء المختلفة
        if not fee:
            if player_name == 'Alexanco':
                fee = TRANSFER_FEES.get(('Alexanko', season))
            elif player_name == 'Schuster':
                fee = TRANSFER_FEES.get(('Bernhard Schuster', season))
        
        player_dict['transfer_fee'] = fee if fee else '0'

# حفظ الملف الجديد
with open('season_players_with_fees.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("تم إضافة الأسعار وحفظ الملف! ✅")
```

```python
import pandas as pd

# 1. تحديد مسارات الملفات بالترتيب اللي إنت عاوزه
files = [
    'barca_80s_updated.xlsx',
    'barca_90s_updated.xlsx',
    'barca_99_10_updated.xlsx',
    'barca_10_25_updated.xlsx'
]

# 2. قراءة الملفات وتخزينها في قائمة
all_dfs = []
for file in files:
    df = pd.read_excel(file)
    print(f"Reading {file}... Rows: {len(df)}")
    all_dfs.append(df)

# 3. دمج الملفات كلها تحت بعض (Vertical Concatenation)
# ignore_index=True عشان يعيد ترتيب الأرقام الجانبية (1, 2, 3...) من جديد للملف الكبير
final_df = pd.concat(all_dfs, ignore_index=True)

# 4. حفظ الملف النهائي الشامل
final_df.to_excel('BARCA_FULL_HISTORY_1980_2026.xlsx', index=False)

print("\n✅ تم دمج الأربعة ملفات بنجاح في ملف واحد!")
print(f"إجمالي عدد اللاعبين في التاريخ: {len(final_df)}")
```
